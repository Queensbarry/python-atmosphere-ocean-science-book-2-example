计算观测站某天逐小时 2m 气温（℃）和边界层高度（m）的皮尔逊相关系数

```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [
    6.3, 5.8, 5.3, 4.7, 4.2, 3.7, 3.3, 2.7, 5.3, 8.5,
    11.3, 11.8, 11.1, 10.3, 9.9, 9.7, 8.8, 7.6, 6.3, 4.8,
    3.6, 2.7, 2.1, 1.6
]
b = [
    270.3, 298.4, 297.2, 274.4, 189.1, 137.4, 126.4, 120.2, 143.9, 103.4,
    327.5, 686.4, 786.4, 825.9, 1011.5, 1004.2, 926.4, 845.1, 766.1, 720.0,
    644.8, 622.4, 606.7, 569.4
]
```

```python
sp.stats.pearsonr(a, b)
```

由上述结果可得出 $pvalue < 0.05$，结果是显著的

```python
a = np.asarray(a)
b = np.asarray(b)

fig = plt.figure(figsize=(10, 5))
plt.plot(range(a.size), a / np.max(np.abs(a)))
plt.plot(range(b.size), b / np.max(np.abs(b)))
```
