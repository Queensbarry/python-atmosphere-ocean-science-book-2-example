计算 1981 年 1 月至 2019 年 12 月 NINO3 指数落后 3 个月的自协方差

```python
import pandas as pd
from datetime import datetime
from qb.statistics.variance_analysis import auto_cov
```

```python
# https://ftp.cpc.ncep.noaa.gov/wd52dg/data/indices/sstoi.indices
DATA_PATH = './data/sstoi.indices'
```

```python
with open(DATA_PATH, 'r+') as f:
    old = f.read()
    if old[: 2] != '  ':
        f.seek(0)
        f.write('  ')
        f.write(old)
```

```python
df = pd.read_fwf(DATA_PATH)
```

```python
df
```

```python
# 设定 DatetimeIndex
df.index = df.apply(lambda x: datetime(int(x[0]), int(x[1]), 1), axis=1)
df.drop(['YR', 'MON'], axis=1, inplace=True)
```

```python
df.index
```

```python
nino3 = df[datetime(1981, 1, 1): datetime(2019, 12, 31)]['NINO3'].to_numpy()
```

```python
nino3
```

```python
auto_cov(nino3, tau=3, bias=True)
```
